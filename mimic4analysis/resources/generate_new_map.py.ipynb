```python
import pandas as pd
import numpy as np
import os
import csv

from tqdm import tqdm


def dataframe_from_csv(path, header=0, index_col=0):
    return pd.read_csv(path, header=header, index_col=index_col)
```

```python
itemids = set()
def read_events_table_by_row(mimic4_path, table):
    nb_rows = {'chartevents': 329499789, 'labevents': 122103668, 'outputevents': 4457382}
    print(os.path.join(mimic4_path, table.lower() + '.csv'))
    reader = csv.DictReader(open(os.path.join(mimic4_path, table.lower() + '.csv'), 'r'))
    for i, row in enumerate(reader):
        if 'stay_id' not in row:
            row['stay_id'] = ''
        yield row, i, nb_rows[table.lower()]
        
def read_the_table(mimic4_path, table, nb_rows = 100000):
    for row, row_no, _ in tqdm(read_events_table_by_row(mimic4_path, table), total=nb_rows,
                                                        desc='Processing {} table'.format(table)):
        itemid = row['itemid']
        itemids.add(itemid)
        
        
        
    # finish traverse the table
    
mimic4_path_chartevents = '/Users/xutian/Documents/Dev/dartmouth/mimic-data/mimic-iv-1.0/icu'
table_chartevents = 'chartevents'
nb_rows_chartevents = 329499789

read_the_table(mimic4_path_chartevents, table_chartevents, nb_rows_chartevents)





```

```python

mimic4_path_labevents = '/Users/xutian/Documents/Dev/dartmouth/mimic-data/mimic-iv-1.0/hosp'
table_labevents = 'labevents'
nb_rows_labevent = 122103668

read_the_table(mimic4_path_labevents, table_labevents, nb_rows_labevent)

```
