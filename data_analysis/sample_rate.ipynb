```python
subject_id_episode_ts_set = set()

invasive_path_train = os.path.join(invasive_path, 'train')

invasive_path_train_listfile = os.path.join(invasive_path_train, 'listfile.csv')

with open(invasive_path_train_listfile, 'r') as ivf:
    headers = ivf.readline()
    for line in ivf:
        
        tokens = line.split(",")
        subject_id_episode_ts_set.add(tokens[0])
        '''
        episode_tokens = tokens[0].split("_")
        subject_id = episode_tokens[0]
        episode = episode_tokens[1]
        subject_id_episode = subject_id + "_" + episode
        subject_id_episode_set.add(subject_id_episode)

        '''
        

print(len(subject_id_episode_ts_set))






    
```

```python
user_feature_summary = {}


headers_general = []
cnt = 0
for s_e_ts in subject_id_episode_ts_set:
    if s_e_ts in ['19902080_episode1_timeseries.csv']:
        continue
    cnt += 1
    if cnt >= 50:
        break
    if user_feature_summary.get(s_e_ts) is None:
        user_feature_summary[s_e_ts] = {
            'los': 0,
            'features_cnt': [],
            'features_feq_24h': []
        }
    ts = pd.read_csv(os.path.join(invasive_path_train, s_e_ts))
    headers = list(ts.columns)
    data = ts.values
    headers_general = headers
    times = data[:, 0]
    last_ts = int(times[len(times) - 1])
    user_feature_summary.get(s_e_ts)['los'] = last_ts
    
    for j in range(1, data.shape[1], 1):
        if j in [17, 18, 19, 27]:
            col = data[:, j]
            vals = [x for x in col if isinstance(x, str)]
        else:
            col = data[:, j]
            print(s_e_ts, j, headers[j], cnt)
            vals = [x for x in col if not np.isnan(x)]
        
        user_feature_summary.get(s_e_ts)['features_cnt'].append(len(vals))
        user_feature_summary.get(s_e_ts)['features_feq_24h'].append(len(vals) * 24. / last_ts)
        

# data = pd.read_csv('18881064_episode1_timeseries.csv')
# print(data)
```

```python

headers_g = headers_general[1:]
print(headers_g)

collect_freq_24_hour = []
for idx, val in user_feature_summary.items():
    features_feq_24h = val['features_feq_24h']
    collect_freq_24_hour.append(features_feq_24h)

collect_freq_24_hour_np = np.array(collect_freq_24_hour)
print(collect_freq_24_hour_np.shape)
max_col = collect_freq_24_hour_np.max(axis = 0)
print(max_col.shape)
print(max_col)

result = []
for i in range(len(headers_g)):
#     print(headers_g[i], max_col[i])
    result.append((headers_g[i], max_col[i]))
    

    
```
